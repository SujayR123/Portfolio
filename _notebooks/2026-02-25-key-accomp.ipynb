## Sprint 2 ‚Äî Learning the Language of Code

Sprint 2 was a deep dive into core programming concepts. Here's everything I covered:

**Iterations** ‚Äî Repeated actions using loops, automating tasks that would be tedious to write manually.

**Conditionals & Nested Conditionals** ‚Äî Logic that lets a program make decisions based on true/false checks. Nesting them unlocks more complex decision paths.

**Arrays** ‚Äî Ordered lists that store multiple values under one variable. Each item has an index for quick access.

**Mathematical Expressions** ‚Äî Using operators like `+`, `-`, `*`, `/` to calculate and evaluate formulas inside a program.

**Booleans** ‚Äî The simplest data type: `true` or `false`. Essential for controlling logic and conditions.

**JSON** ‚Äî A lightweight format for storing and sharing structured data using key-value pairs and arrays.

**Data Abstraction** ‚Äî Hiding complex details and only exposing what's necessary, making code easier to read and maintain.

**Strings** ‚Äî Sequences of characters for representing text. Can be combined, searched, sliced, and formatted.

**Classes & Methods** ‚Äî Classes are blueprints for objects; methods are their behaviors. The backbone of OOP and reusable code.

**Variables & the HTML DOM** ‚Äî Variables store data JavaScript can use, and the DOM represents a webpage as a structured element tree. Together they let code read and update live webpage content.

Beyond the concepts, I also got hands-on with **Jupyter Notebooks**, learning about:
- Permalinks and frontmatter
- Markdown and JavaScript cells
- The CodeRunner
- Submitting **pull requests** to publish my own lesson

## Sprint 3 ‚Äî Building an RPG Game

For Sprint 3, my group built a full **RPG game** using the GameRunner and GameEngine framework.

I served as a **lead coder** and was responsible for:
- Integrating the **background** into the level
- Building and placing **NPCs**
- Implementing **WASD movement** for the player

The biggest thing I learned was how **game architecture** is structured ‚Äî levels are defined as classes with constructors that hold the background, player, NPCs, and all other game objects in an array. Understanding that pattern made it clear how large codebases can stay organized even as they grow in complexity.
